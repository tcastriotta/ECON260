# Q1

```python
q1, q2, p1, p2, y, Ubar = symbols('q1 q2 p1 p2 y \overline{U}')

alpha = symbols('alpha')
```

a) If preference is represented by: 

\begin{align}
U(q_1,q_2) &= q_1^.4q_2^.6
\end{align}

The consumer's optimal choice problem that minimizes her expenditure would be:

\begin{align}
  minC_{q_1,q_2}(q_1 , q_2) = p_1q_1^.4 + p_2q_2^.6\\
   s.t. \overline{U} = q_1^.4q_2^.6 \\
\end{align}

Thw choice variables would be the $q_1$ and $q_2$ under "min" , 
The objective function is the first equation in its entirety ,
he constraint function is the second equation that equates to $\overline{U}$, or max utility level 

b) Lagrangian Would Be:
    
\begin{align}
    L = p_1q_1^.4 + p_2q_2^.6 + \lambda(\overline{U} - q_1^.4q_2^.6)
\end{align}

```python
U = p1*q1**(.4) + p2*q2**(.6)

U
```

```python
lam = symbols('lambda', real = True, Positive = True)
```

```python
lagr = U + lam * (Ubar - q1**.4 * q2**.6)

lagr
```

c) F.O.C.'s

(1)

\begin{align}
\frac{\partial L}{\partial q_1} &= p_1 -\lambda.4q_1^{-.6}q_2^.6 = 0 \\
    &= \lambda.4q_1^{-.6}q_2^.6 = p_1 \\
\end{align}

```python
lagr.diff(q1)
```

(2)

\begin{align}
\frac{\partial L}{\partial q_2} &= p_2 - \lambda(.6)q_1^.4q_2^{-.4} = 0 \\
    &= \lambda(.6)q_1^.4q_2^{-.4} = p_2 \\
\end{align}

```python
lagr.diff(q2)
```

(3)

\begin{align}
\frac{\partial L}{\partial \lambda} &= \overline{U} - q_1^.4q_2^.6 = 0 \\
\end{align}

```python
lagr.diff(lam)
```

d) Solving for optimal consumption bundles

(4) Using (1) and (2) to find MRS = MRT:

\begin{align}
\frac{.4}{.6} \frac{q_2}{q_1} = \frac{p_1}{p_2} \\
    &= \frac{.6}{.4} \frac{p_1}{p_2} q_1 \\
\end{align}

Plugging (4) into (3)

\begin{align}
\overline{U} - q_1^.4(\frac{.6}{.4} \frac{p_1}{p_2} q_1)^.6 \\
    &= (\frac{.6}{.4} \frac{p_1}{p_2})^.6 q_1 = \overline{U} \\
    &= q_1^* = (\frac{.6}{.4} \frac{p_1}{p_2})^{-.6} \overline{U} \\
\end{align}

Plug $q_1^*$ into (4)

\begin{align}
q_2 = \frac{.6}{.4} \frac{p_1}{p_2}(\frac{.6}{.4} \frac{p_1}{p_2})^{-.6} \overline{U} \\
    &= q_2^* = (\frac{.6}{.4} \frac{p_1}{p_2})^.4 \overline{U} \\
\end{align}

Optimal Consumption Bundles:

\begin{align}
(q_1^* = (\frac{.6}{.4} \frac{p_1}{p_2})^{-.6} \overline{U} , q_2^* = (\frac{.6}{.4} \frac{p_1}{p_2})^.4 \overline{U})) \\
\end{align}


```python
foc_sol = solve( ( diff(lagr,q1), diff(lagr,q2), diff(lagr,lam) ), q1,q2,lam)

foc_sol
```
